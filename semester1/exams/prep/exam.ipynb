## Assignment 7: Network Metrics

### Directed Graph

```python
DiG = nx.DiGraph()
DiG.add_edges_from([(1,2),(1,3),(2,3),(2,4),(3,1)])

nx.draw_circular(DiG,with_labels=True)

# basic network metrics
print(f"N: {len(DiG.nodes())}")
print(f"L: {len(DiG.edges())}\n")

# degrees and average degree (for both in and out degree)
out_degrees = {node: out_degree for node,out_degree in DiG.out_degree()}
in_degrees = {node: in_degree for node, in_degree in DiG.in_degree()}
degrees  = {node: degree for node, degree in DiG.degree()}
print(out_degrees)
print(f"Out-Degrees: {out_degrees} (Mean: {np.mean([i for i in out_degrees.values()])})")
print(f"In-Degrees: {in_degrees} (Mean: {np.mean([i for i in in_degrees.values()])})")
print(f"Degrees: {degrees}")
print(f"Total Average Degree: {np.mean([i[1] for i in DiG.degree()])}\n")

# average shortest paths
print(f"Average Shortest Paths: {nx.average_shortest_path_length(DiG)}")
try: 
    print(f"Diameter: {nx.diameter(DiG)}\n")
except: None

# local clustering coefficient
print(f"Clustering Coefficients: {nx.clustering(DiG)}\n")

# centralities
print(f"Degree Centralities: {nx.degree_centrality(DiG)}")
print(f"Betweenness Centralities: {nx.betweenness_centrality(DiG)}")
print(f"Closeness Centralities: {nx.closeness_centrality(DiG)}")
```

### Undirected Graph

```python
# enter the edges
edge_list = [(1,2),(1,6),(2,3),(2,6),(3,4),(5,6)]
edge_list2 = [(1,2),(1,3),(1,4),(2,4),(2,5),(3,5),(4,5),(4,6)]

# instantiate the graph nx.Graph object (for directed: nx.DiGraph())
G = nx.from_edgelist(edge_list2)

#########################
# calculate some important metrics
N = len(G.nodes()) # equivalent to len(G)
L = len(G.edges()) # equivalent to G.size()
print(f"Number of Nodes (N): {N}")
print(f"Number of Edges (L): {L}")

# degrees
degrees = {node: degree for node, degree in G.degree()}
print(f"Degrees: {degrees}")

min_degree = min(degrees.values())
max_degree = max(degrees.values())

average_degree = sum(degrees.values())/(N-1)
print(f"Average Degree: {average_degree}")
print(f"Minimum Degree: {min_degree}")
print(f"Maximum Degree: {max_degree}\n")

# average shortest paths
print(f"Average Shortest Paths: {nx.average_shortest_path_length(G)}")

# diameter
print(f"Diameter: {nx.diameter(G)}\n")

# clustering coefficients
print(f"Clustering Coefficients: {nx.clustering(G)}\n")

# degree centrality 
print(f"Degree Centralities: {nx.degree_centrality(G)}")

# betweenness centrality
print(f"Betweenness Centralities: {nx.betweenness_centrality(G,normalized=False)}")

# closeness centrality
print(f"Closeness Centralities: {nx.closeness_centrality(G)}")
############################

nx.draw(G, with_labels=True)
```
